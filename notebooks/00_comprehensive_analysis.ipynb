```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('default')
sns.set_palette("husl")

print("📦 All libraries loaded successfully!")
print("🎯 Ready for comprehensive analysis!")
```

## 📊 REQUIREMENT 1: Statistical Plots from Excel Data

```python
# Load data from CSV (Excel equivalent)
df = pd.read_csv('../data/processed/cities_data.csv')

print(f"📊 Dataset Loaded Successfully!")
print(f"📋 Dataset Shape: {df.shape}")
print(f"🏙️ Cities: {len(df)}")
print(f"📈 Metrics: {len(df.columns)}")

# Display first few rows
print("\n📋 Sample Data:")
df.head()
```

```python
# Create histogram for UGI distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['UGI'], bins=15, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribution of UGI Scores', fontsize=14, fontweight='bold')
plt.xlabel('UGI Score')
plt.ylabel('Number of Cities')
plt.grid(True, alpha=0.3)

# Population histogram
plt.subplot(1, 2, 2)
plt.hist(df['Population_2011'], bins=12, alpha=0.7, color='lightcoral', edgecolor='black')
plt.title('Population Distribution (2011 Census)', fontsize=14, fontweight='bold')
plt.xlabel('Population (Millions)')
plt.ylabel('Number of Cities')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ Histograms created successfully!")
```

```python
# Create pie charts
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Pie Chart 1: Cities by Region
region_counts = df['Region'].value_counts()
axes[0].pie(region_counts.values, labels=region_counts.index, autopct='%1.1f%%', startangle=90)
axes[0].set_title('Cities Distribution by Region', fontsize=14, fontweight='bold')

# Pie Chart 2: Metro vs Non-Metro Cities
metro_counts = df['Metro'].value_counts()
metro_labels = ['Non-Metro Cities', 'Metro Cities']
axes[1].pie(metro_counts.values, labels=metro_labels, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Metro vs Non-Metro Cities', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print("✅ Pie charts created successfully!")
```

```python
# Create bar charts
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Bar Chart 1: Top 10 Cities by UGI
top_10_ugi = df.nlargest(10, 'UGI')
axes[0,0].barh(top_10_ugi['City'], top_10_ugi['UGI'], color='skyblue')
axes[0,0].set_title('Top 10 Cities by UGI Score', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('UGI Score')

# Bar Chart 2: Top 10 Cities by Population
top_10_pop = df.nlargest(10, 'Population_2011')
axes[0,1].barh(top_10_pop['City'], top_10_pop['Population_2011'], color='lightcoral')
axes[0,1].set_title('Top 10 Cities by Population', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Population (Millions)')

# Bar Chart 3: Top 10 Cities by GDP
top_10_gdp = df.nlargest(10, 'GDP_Billions_USD')
axes[1,0].barh(top_10_gdp['City'], top_10_gdp['GDP_Billions_USD'], color='lightgreen')
axes[1,0].set_title('Top 10 Cities by GDP', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('GDP (Billions USD)')

# Bar Chart 4: Average UGI by Region
regional_ugi = df.groupby('Region')['UGI'].mean().sort_values(ascending=True)
axes[1,1].barh(regional_ugi.index, regional_ugi.values, color='gold')
axes[1,1].set_title('Average UGI Score by Region', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Average UGI Score')

plt.tight_layout()
plt.show()

print("✅ Bar charts created successfully!")
```

## 🗺️ REQUIREMENT 2: Geographic Visualizations on India Map

```python
# Map 1: Comprehensive UGI Map of India
fig_map1 = px.scatter_geo(df, 
                         lat='Latitude', lon='Longitude',
                         size='UGI', color='UGI',
                         hover_name='City',
                         hover_data=['State', 'Population_2011', 'GDP_Billions_USD'],
                         title='🗺️ India Urban Growth Intelligence Map',
                         color_continuous_scale='RdYlGn',
                         size_max=40)

fig_map1.update_geos(
    projection_type="natural earth",
    showland=True, landcolor="lightgray",
    showocean=True, oceancolor="lightblue",
    center_lat=20, center_lon=78,
    projection_scale=3
)

fig_map1.show()
print("🗺️ Interactive UGI Map created!")
```

```python
# Map 2: Population Distribution Map
fig_map2 = px.scatter_geo(df,
                         lat='Latitude', lon='Longitude',
                         size='Population_2011', color='Population_2011',
                         hover_name='City',
                         title='👥 Population Distribution Across India',
                         color_continuous_scale='Blues',
                         size_max=35)

fig_map2.update_geos(
    projection_type="natural earth",
    showland=True, landcolor="lightgray",
    center_lat=20, center_lon=78,
    projection_scale=3
)

fig_map2.show()
print("👥 Population Map created!")
```

## 💡 REQUIREMENT 3: Data-Driven Insights and Recommendations

```python
# Generate comprehensive insights
print("🎯 COMPREHENSIVE DATA INSIGHTS")
print("=" * 50)

# Overall Statistics
print("\n📊 OVERALL STATISTICS:")
print(f"• Total cities analyzed: {len(df)}")
print(f"• Average UGI Score: {df['UGI'].mean():.1f}")
print(f"• Total urban population: {df['Population_2011'].sum():.1f} million")
print(f"• Combined GDP: ${df['GDP_Billions_USD'].sum():.1f} billion")

# Top Performers
print("\n🏆 TOP PERFORMERS:")
top_5 = df.nlargest(5, 'UGI')
for i, (_, city) in enumerate(top_5.iterrows(), 1):
    print(f"{i}. {city['City']}, {city['State']} - UGI: {city['UGI']:.1f}")

# Regional Analysis
print("\n🗺️ REGIONAL ANALYSIS:")
regional_stats = df.groupby('Region').agg({
    'UGI': 'mean',
    'Population_2011': 'sum',
    'GDP_Billions_USD': 'sum',
    'City': 'count'
}).round(2)

for region in regional_stats.index:
    stats = regional_stats.loc[region]
    print(f"• {region}: {stats['City']} cities, Avg UGI: {stats['UGI']:.1f}")

# Infrastructure Analysis
print("\n🏗️ INFRASTRUCTURE INSIGHTS:")
metro_avg = df[df['Metro'] == 1]['UGI'].mean()
non_metro_avg = df[df['Metro'] == 0]['UGI'].mean()
print(f"• Metro cities average UGI: {metro_avg:.1f}")
print(f"• Non-metro cities average UGI: {non_metro_avg:.1f}")
print(f"• Metro advantage: {((metro_avg/non_metro_avg - 1) * 100):.1f}% higher UGI")
```

```python
# Strategic Recommendations
print("\n🎯 STRATEGIC RECOMMENDATIONS")
print("=" * 50)

print("\n📈 POLICY RECOMMENDATIONS:")
print("1. 🎯 Focus on Tier-2 cities with UGI 50-80 for balanced development")
print("2. ✈️ Prioritize airport connectivity for cities with >1.5M population")
print("3. 🚇 Expand metro systems in high-population, high-UGI cities")
print("4. 🌐 Strengthen digital infrastructure in emerging regions")

print("\n⚠️ KEY CHALLENGES:")
print("1. 📊 Urban inequality between cities and regions")
print("2. 🗺️ Regional imbalance in development")
print("3. 🚧 Infrastructure gaps in smaller cities")
print("4. 🌱 Need for sustainable urban development")

print("\n✅ SUCCESS FACTORS:")
print("1. 🔗 Strong correlation between connectivity and development")
print("2. 🏙️ Metro systems significantly boost city performance")
print("3. ✈️ Airport connectivity crucial for economic growth")
print("4. 🎯 Balanced approach needed for sustainable development")
```
